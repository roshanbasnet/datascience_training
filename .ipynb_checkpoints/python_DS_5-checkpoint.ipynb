```python
columns_to_drop = ["PassengerId", "Cabin", "Name"]
```

```python
data = data.drop(columns_to_drop, axis=1)
```

```python
data = data.drop("Ticket", axis=1)
```

```python
data.head(1)
```

```python
sns.countplot("Pclass", data=data, hue="Survived")
plt.show()
```

```python
sns.countplot("Sex", data=data, hue="Survived")
plt.show()
```

```python
sns.countplot("Embarked", data=data, hue="Survived")
plt.show()
```

```python
data["Family"] = data["SibSp"] + data["Parch"]
```

```python
data.drop(['SibSp', 'Parch'], inplace=True, axis=1)
```

```python
sns.countplot("Family", data=data, hue="Survived")
plt.show()
```

```python
data.Embarked.value_counts()
```

```python
data["Embarked"].fillna("S", inplace=True) 
```

```python
plt.hist(data["Age"], bins=100)
plt.show()
```

```python
low = data.Age.mean() - data.Age.std()
high = data.Age.mean() + data.Age.std()

random_age = lambda : int(np.random.randint(low=low, high=high, size=1))
```

```python
data.Age.fillna(random_age(), inplace=True)
```

```python
data.isna().sum()
```

```python
data["Sex"] = pd.Categorical(data["Sex"])
data["Embarked"] = pd.Categorical(data["Embarked"])
```

```python
data = pd.get_dummies(data)
```
