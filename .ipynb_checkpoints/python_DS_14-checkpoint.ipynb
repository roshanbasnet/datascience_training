```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()
X = iris_data.data
y = iris_data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
dtree = DecisionTreeClassifier(criterion='entropy')
```

```python
dtree.fit(X_train, y_train)
```

```python
dtree.feature_importances_
```

```python
y_pred = dtree.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
plt.figure(figsize=(18,18))
plot_tree(dtree, 
          class_names=iris_data.target_names,
         feature_names= iris_data.feature_names)
plt.show()
```

```python
rnd_forest = RandomForestClassifier(n_estimators=150, criterion='gini')
```

```python
rnd_forest.fit(X_train, y_train)
y_pred = rnd_forest.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
rnd_forest.feature_importances_
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
estiimators = [("KNN", knn), ("DecisionTree", dtree), 
               ("RandomForest", rnd_forest)]

ensemble_model = VotingClassifier(estimators=estiimators, 
                                  weights=[0.33, 0.33, 0.33])

ensemble_model.fit(X_train, y_train)
y_pred = ensemble_model.predict(X_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
stomp = DecisionTreeClassifier(max_depth=1)
```

```python
ada_boost = AdaBoostClassifier(n_estimators=100, base_estimator=stomp)
```

```python
ada_boost.fit(X_train, y_train)
y_pred = ada_boost.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
ada_boost.feature_importances_
```
