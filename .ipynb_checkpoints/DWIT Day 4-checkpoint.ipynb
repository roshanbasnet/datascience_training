```python
from typing import List
```

```python
def sum_calc_list(list1:List) ->int:
    """
    Returns sum of a list
    Paramaters:
    list1:list
    Returns:
    sum_list: Int
    """
    sum_list = 0
    for item in list1:
        sum_list += item
    return sum_list
```

```python
sum_calc_list([1,2,3,4])
```

```python
def sum_calc_list(list1:List) ->int:
    """
    Returns sum of a list
    Paramaters:
    list1:list
    Returns:
    sum_list: Int
    """
    sum_list = 0
    for item in list1: #[1,2,3,[4,5]]
        if type(item) == type([]):
            for i in item: #[4,5]
                sum_list += i
        else:
            sum_list += item
    return sum_list
```

```python
sum_calc_list([1,2,3,[4,5]])
```
