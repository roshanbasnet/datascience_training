```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
cat_col = np.array(["Nepal", "Nepal", "India", "China", "India", "Nepal"])
```

```python
label_encoder = LabelEncoder()
```

```python
label_encoder.fit_transform(cat_col.reshape(-1, 1))
```

```python
label_encoder.inverse_transform([2, 2, 1, 0, 1, 2])
```

```python
onehot_encoder = OneHotEncoder(sparse=False)
```

```python
onehot_encoder.fit_transform(cat_col.reshape(-1, 1))
```

```python
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# plt.figure(figsize=(18, 10))
plt.title("X vs Sin(X) vs Cos(X)", {"fontsize": 20})
plt.xlabel("X")
plt.plot(x, y1, color="black", label="Sin(X)", linestyle=":")
plt.plot(x, y2, color="green", label="Cos(X)")
plt.legend()
plt.show()
```

```python
x_cols= ["Nepal", "India", "China"]
y_vals = [100, 56, 67]
```

```python
plt.bar(x_cols, y_vals)
plt.show()
```

```python
plt.pie(x=y_vals, labels=x_cols, explode=[0.05, 0, 0])
plt.show()
```
