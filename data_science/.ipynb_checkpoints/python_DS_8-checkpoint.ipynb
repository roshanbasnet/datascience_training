```python
h, w, ch = image.shape
```

```python
image = image.reshape(h*w, ch)
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(image)
```

```python
kmeans.labels_
```

```python
data = np.array([
    [1,2],
    [2,4],
    [3,4],
    [1,5],
    [4,5]
])
```

```python
plt.scatter(x=data[:,0], y=data[:,1])
plt.show()
```

```python
meu = data.mean(axis=0)
std = data.std(axis=0)
```

```python
data = (data - meu) / std
```

```python
sigma = np.cov(data.T)
```

```python
sigma
```

```python
eig_val, eig_vec = np.linalg.eig(sigma)
```

```python
max_index = eig_val.argmax()
```

```python
max_index
```

```python
axis1 = eig_vec.T[max_index]
```

```python
new_points = np.dot(data, axis1)
```

```python
new_points.shape
```

```python
plt.scatter(x=new_points, y=np.zeros(5))
plt.show()
```
