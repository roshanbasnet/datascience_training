```python
# SVM: Support Vector Machine
# also maximum margin classifier.
# take first two support vector.(the vecor are the closest distance from the given class)
# draw the margin from the two support vector

# Hardness of SVM:strciltly tries to classify the given data into two class. It does not work 
#     properly if there isoutlier
# softness of SVM: leav the misclassify point as it is and try to make the general margin maximum
# https://medium.com/machine-learning-101/chapter-2-svm-support-vector-machine-theory-f0812effc72
# We can use rbf function to translate the linear function into higher dimension
# https://towardsdatascience.com/support-vector-machines-svm-c9ef22815589
# http://web.mit.edu/6.034/wwwbob/svm.pdf
```

```python

import numpy as np
import pandas as pd
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_circles
```

```python
from sklearn.svm import LinearSVC, SVC
```

```python
# give x and y
X, y = make_circles(n_samples=200,noise=0.1,factor=0.6)
```

```python
# X alays have two dimension
y
```

```python
plt.scatter(X[y==0,0],X[y==0,1],c='r')
plt.scatter(X[y==1,0],X[y==1,1],c='b')
plt.show()
```

```python
lin_svm = LinearSVC()
```

```python
lin_svm.fit(X,y)
```

```python
plot_decision_regions(X,y,lin_svm)
plt.show()
```

```python
# using the linear seperation points are misclassified. So we have to use non linear function
# eg of degree 2 ax2+bx+c
svm = SVC(kernel='poly',degree=2)
svm.fit(X,y)
```

```python
plot_decision_regions(X,y,svm)
plt.show()
```

```python
# Or we can use RBF with out giving degree
svm = SVC(kernel='rbf')
svm.fit(X,y)
```

```python
plot_decision_regions(X,y,svm)
plt.show()
```
